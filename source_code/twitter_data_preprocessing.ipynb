```python
import pandas as pd
from sklearn.model_selection import train_test_split

# -----------------------
# LOAD DATA
# -----------------------

df = pd.read_csv("twitter_data_clean.csv")

# Optional: shuffle entire dataframe first
df = df.sample(frac=1, random_state=42).reset_index(drop=True)

# -----------------------
# FIRST SPLIT
# Train = 2775
# Remaining = 692
# -----------------------

train_df, temp_df = train_test_split(
    df,
    test_size=692,
    stratify=df["sarcastic"],
    random_state=42
)

# -----------------------
# SECOND SPLIT
# Val = 346
# Test = 346
# -----------------------

val_df, test_df = train_test_split(
    temp_df,
    test_size=0.5,
    stratify=temp_df["sarcastic"],
    random_state=42
)

# -----------------------
# SAVE FILES
# -----------------------

train_df.to_csv("train.csv", index=False)
val_df.to_csv("val.csv", index=False)
test_df.to_csv("test.csv", index=False)

# -----------------------
# CHECK SIZES
# -----------------------

print("Train:", len(train_df))
print("Val:", len(val_df))
print("Test:", len(test_df))

# -----------------------
# CHECK LABEL DISTRIBUTION
# -----------------------

print("\nTrain label distribution:")
print(train_df["sarcastic"].value_counts())

print("\nVal label distribution:")
print(val_df["sarcastic"].value_counts())

print("\nTest label distribution:")
print(test_df["sarcastic"].value_counts())
```
